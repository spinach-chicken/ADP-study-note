```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, f1_score, recall_score, precision_score, accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import ColumnTransformer

num_col1 = ['거주개월', '평형대', '월세(원)', '보증금(원)', '대표나이', '나이', '거주자 수']
num_col2 = ['아파트 ID', '아파트 평점', '층', '입주연도', '거주연도']
cat_col = ['계약구분', '성별', '결혼여부']

tf = ColumnTransformer([('num1', StandardScaler(), num_col1), ('num2', MinMaxScaler(), num_col2), ('cat', OneHotEncoder(), cat_col)])
LE = LabelEncoder()

x1 = df_copy_v2.drop(['재계약횟수', '재계약횟수구분'], axis = 1)
y1 = df_copy_v2[['재계약횟수']]
x2 = df_copy_v2.drop(['재계약횟수', '재계약횟수구분'], axis = 1)
y2 = df_copy_v2[['재계약횟수구분']]

x1_trans = tf.fit_transform(x1)
x1_train, x1_test, y1_train, y1_test = train_test_split(x1_trans, y1, test_size = 0.3, random_state = 42)

x2_trans = tf.fit_transform(x2)
y2_trans = LE.fit_transform(y2)
x2_train, x2_test, y2_train, y2_test = train_test_split(x2_trans, y2_trans, test_size = 0.3, random_state = 42, stratify = y2_trans)

model1 = RandomForestRegressor()
model2 = XGBRegressor()
model3 = RandomForestClassifier()
model4 = KNeighborsClassifier()

model1.fit(x1_train, y1_train)
model2.fit(x1_train, y1_train)
model3.fit(x2_train, y2_train)
model4.fit(x2_train, y2_train)

y1_pred1 = model1.predict(x1_test)
y1_pred2 = model2.predict(x1_test)
y2_pred3 = model3.predict(x2_test)
y2_pred4 = model4.predict(x2_test)

print(f'RFR rmse :{mean_squared_error(y1_test, y1_pred1)}')
print(f'XGB rmse :{mean_squared_error(y1_test, y1_pred2)}')
print(f'RFC accuracy : {accuracy_score(y2_test, y2_pred3)}')
print(f'KNN accuracy : {accuracy_score(y2_test, y2_pred4)}')
print('성능이 제일좋은 RandomForestClassifier 선택')
```

## 4-1 각 회차별로 1번 타자의 출루 (1,2,3루타와 사사구(볼넷, 몸에맞는공))가 있는 경우에 대해 득점이 발생 했는지 확인하고자 한다. 이를 위한 전처리를 수행하라. (단, 첫 번째 혹은 두 번째 타자가 홈런을 친 경우 해당 회차 데이터는 제외한다.)     

조건1 : 득점여부를 범주형 종속변수로 한다. (1점이상 득점 :1, 무득점 :0)     
조건2 : 각 회차 2번 타자의 데이터는 원핫 인코딩한다.    
조건3 : 학습에 적절하지 않은 데이터는 제외한다.    

제가 이해하고 의도한 29회 4-1번 문제에 대한 가이드 입니다.
(4-1) 전처리 가이드
1. 첫타자와 둘째타자의 각 이닝별 출루현황, 해당 이닝의 득점수를 3columns의 데이터로 변환
2. 정수형 값이 아닌 데이터 행를 제거
3. 홈런 (4)값을 가지는 행을 제거
4. 이닝별 점수를 binary로 변환 (득점 있으면 1 ,없으면 0 )
5. 둘째 타자 출루 결과를 one-hot encoding

```python
base_data = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/29/p2_.csv')
base_data
```

```python
base_df = pd.DataFrame(columns = ['1번타자행동', '2번타자행동', '득점여부'])
for i in range(1,10):
    concat_df = base_data[['a%d_1'%i, 'a%d_2'%i, 'b%d'%i]]
    concat_df.rename(columns = {'a%d_1'%i:'1번타자행동', 'a%d_2'%i:'2번타자행동', 'b%d'%i:'득점여부'}, inplace =True)
    base_df = pd.concat([base_df,concat_df])

base_df_copy = base_df[base_df['1번타자행동'] != " "]
base_df_copy = base_df_copy[base_df_copy['1번타자행동'] != '"']
base_df_copy = base_df_copy[base_df_copy['1번타자행동'] != ' :']
base_df_copy = base_df_copy[base_df_copy['1번타자행동'] != '&']
base_df_copy = base_df_copy.astype(str)
base_df_copy = base_df_copy[(base_df_copy['1번타자행동'] !='4') & (base_df_copy['2번타자행동'] !='4')]
base_df_copy['득점여부'] = base_df_copy.득점여부.apply(lambda x : 0 if x == "0" else 1)
base_df_onehot = pd.get_dummies(base_df_copy[['2번타자행동']])
base_df_final = pd.concat([base_df_copy[['1번타자행동']], base_df_onehot, base_df_copy[['득점여부']]], axis = 1)
base_df_final
```

## 4-2 4-1 데이터에 대해 Logistic Regression을 적용하고 2번타자의 희생번트 여부에 대한 회귀 계수 검정을 하라

```python
import statsmodels.api as sm

base_df_final['1번타자행동'] = base_df_final['1번타자행동'].astype(int)

x = base_df_final.drop('득점여부', axis = 1)
y = base_df_final[['득점여부']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42, stratify = y)

logit_model = sm.Logit(y_train, sm.add_constant(x_train))

result = logit_model.fit()

print(result.summary())
```

```python
print('2번타자 희생번트 여부의 계수는 -2.796이고 p-value는 1.0으로 귀무가설을 기각하지 못한다.')
print('즉, 분석결과 2번타자 희생번트 여부는 유의하지 않다.')
```

## 4-3 SMOTE (random_state =0 지정)를 적용하여 data imbalance를 해결하라

```python
from imblearn.over_sampling import SMOTE

# SMOTE 객체 생성
smote = SMOTE(random_state=42)

# 소수 클래스 데이터에 적용
x_resampled, y_resampled = smote.fit_resample(x, y)
x_resampled
```

## 6. C사 생산 제품 1000개 중 양품이 600개, D사 생산 제품 500개 중 양품이 200개 이다. 두 회사의 양품률에 차이가 있는지 검정하여라.

```python
from statsmodels.stats.proportion import proportions_ztest

# C사와 D사의 양품 및 불량품 개수
C_good, C_total = 600, 1000
D_good, D_total = 200, 500

# 각 회사의 양품률 계산
p1 = C_good / C_total
p2 = D_good / D_total

# 각 회사의 샘플 크기
n1 = C_total
n2 = D_total

# proportions_ztest 함수로 Z 점수와 p-value 계산
Z, p_value = proportions_ztest([C_good, D_good], [C_total, D_total])

# 유의수준 설정 (보통 0.05 사용)
alpha = 0.05

# 가설 검정 결과 출력
if p_value < alpha:
    print("귀무가설을 기각합니다. 두 회사의 양품률에는 통계적으로 유의한 차이가 있습니다.")
else:
    print("귀무가설을 기각하지 못합니다. 두 회사의 양품률에는 통계적으로 유의한 차이가 없습니다.")

# Z 점수와 p-value 출력
print("Z 점수:", Z)
print("p-value:", p_value)
```

### 7-1. 각 차종 별 범퍼 파손의 정도에 차이가 유의한지 검정하라.

```python
data = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/29/p7.csv')
data
```

```python
from scipy.stats import f_oneway


# ANOVA 검정 실행
statistic, p_value = f_oneway(*[group["ratio"] for name, group in data.groupby("name")])

# 유의수준 설정 (보통 0.05 사용)
alpha = 0.05

# p-value를 기준으로 귀무 가설 기각 여부 결정
if p_value < alpha:
    print("귀무가설을 기각합니다. 차종별로 범퍼 파손정도에 통계적으로 유의한 차이가 있습니다.")
else:
    print("귀무가설을 기각하지 못합니다. 차종별로 범퍼 파손정도에 통계적으로 유의한 차이가 없습니다.")

# 검정 통계량과 p-value 출력
print("검정 통계량:", statistic)
print("p-value:", p_value)
```

### 7-2 귀무가설을 채택한다면 그 의미를 해석하고, 귀무가설을 기각하였다면 사후분석을 시행하라.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd


# Tukey HSD 분석 실행
tukey_results = pairwise_tukeyhsd(data['ratio'], data['name'])

# 결과 출력
print(tukey_results)
```
